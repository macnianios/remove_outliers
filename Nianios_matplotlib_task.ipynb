# **Task 1**

Create a scatter plot using the Iris dataset to visualize the relationship between the sepal length and sepal width of the flowers per specie.
Instructions: The plot should contain all the necessary components (labels, title, etc.). Try to build the plot step by step

**Analysis**: What observations can you make regarding the relationship between sepal length and sepal width among the different species of iris flowers

**Bonus**: Experiment with other arguments of the scatter function.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
# extract data for each species
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']
```

```python
#reset figure style
plt.style.use("default")
```

```python
#'built basic scatter plot'
plt.scatter(data=iris, x='sepal_length',y='sepal_width')
plt.show()
```

```python
plt.scatter(data=iris, x='sepal_length',y='sepal_width')
plt.xlabel("Sepal Length") # add xlabel
plt.ylabel("Sepal Width") # add ylabel
plt.title("Sepal Width vs Sepal Length") # add plot title
plt.show()
```

```python
#build three scatter plots, ever plot with unique default color
plt.scatter(data=setosa, x='sepal_length',y='sepal_width',label="setosa")
plt.scatter(data=versicolor, x='sepal_length',y='sepal_width',label="versicolor")
plt.scatter(data=virginica, x='sepal_length',y='sepal_width',label="virginica")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Width vs Sepal Length")
plt.legend() #add legend
plt.show()
```

```python
plt.figure(figsize=(16,8)) #change the figure size

plt.scatter(data=setosa, x='sepal_length',y='sepal_width',label="setosa")
plt.scatter(data=versicolor, x='sepal_length',y='sepal_width',label="versicolor")
plt.scatter(data=virginica, x='sepal_length',y='sepal_width',label="virginica")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Width vs Sepal Length")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,8))

#set new specific color for each species
plt.scatter(data=setosa, x='sepal_length',y='sepal_width',label="setosa",color="brown")
plt.scatter(data=versicolor, x='sepal_length',y='sepal_width',label="versicolor",color="deeppink")
plt.scatter(data=virginica, x='sepal_length',y='sepal_width',label="virginica",color="lightsalmon")

#change label text properties, font size,style and weight
plt.xlabel("Sepal Length", fontsize = 12, fontstyle="italic")
plt.ylabel("Sepal Width",fontsize = 12,fontstyle="italic")
plt.title("Sepal Width vs Sepal Length",fontsize = 18, fontweight = 'bold')
plt.legend(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16,8))

# Create a ListedColormap from colormap and set new colors for species
from matplotlib.colors import ListedColormap
cmap = ListedColormap(["darkorange", "gold", "lawngreen"])

plt.scatter(data=setosa, x='sepal_length',y='sepal_width',label="setosa",color=cmap(0))
plt.scatter(data=versicolor, x='sepal_length',y='sepal_width',label="versicolor",color=cmap(1))
plt.scatter(data=virginica, x='sepal_length',y='sepal_width',label="virginica",color=cmap(2))


plt.xlabel("Sepal Length", fontsize = 12, fontstyle="italic")
plt.ylabel("Sepal Width",fontsize = 12,fontstyle="italic")
plt.title("Sepal Width vs Sepal Length",fontsize = 18, fontweight = 'bold')
plt.legend(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16,8))

from matplotlib.colors import ListedColormap
cmap = ListedColormap(["darkorange", "gold", "purple"])

#change markers color,shape and  size so we can detect the overlapping samples
plt.scatter(data=setosa, x='sepal_length',y='sepal_width',label="setosa",color=cmap(0),marker=(5, 0),linewidths=6)
plt.scatter(data=versicolor, x='sepal_length',y='sepal_width',label="versicolor",color=cmap(1),marker=(6, 0),linewidths=9)
plt.scatter(data=virginica, x='sepal_length',y='sepal_width',label="virginica",color=cmap(2),marker=(6, 2),linewidths=3)


plt.xlabel("Sepal Length", fontsize = 12, fontstyle="italic")
plt.ylabel("Sepal Width",fontsize = 12,fontstyle="italic")
plt.title("Sepal Width vs Sepal Length",fontsize = 18, fontweight = 'bold')
plt.legend(fontsize=14)
plt.show()
```

# **ANALYSIS**

This plot clearly shows that "setosa" flowers form a distinct group on the right. Sepal length for setosas increases as sepal width increases, indicating a positive correlation. In contrast, virginica flowers tend to have a larger sepal length than versicolor, with a slight difference in width. Interestingly, there are ten data points where versicolor and virginica flowers overlap in terms of both sepal length and width. This suggests the need for additional characteristics or bigger sample size to differentiate these overlapping samples more effectively.
